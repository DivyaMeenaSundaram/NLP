```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
import re
```

```python
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
text = "Dr. John Smith travelled from Chennai to Bengaluru to present his research paper."

print("Original Text:")
print(text)
```

```python
tokens = word_tokenize(text)
print("\nStep 1: Tokenization")
print(tokens)
```

```python
text_lower = text.lower()
```

```python
clean_text = re.sub(r'[^a-zA-Z\s]', '', text_lower)
```

```python
clean_tokens = word_tokenize(clean_text)

print("\nStep 2: Preprocessing")
print(clean_tokens)
```

```python
print("\nStep 3: Loading Language Model")
print("NLTK pre-trained POS tagger is ready.")
```

```python
print("\nStep 4: Linguistic Analysis")

for word in clean_tokens:
    print(word)
```

```python
pos_tags = pos_tag(clean_tokens)

print("\nStep 5: POS Tagging")

for word, tag in pos_tags:
    print(word, "->", tag)
```

```python
print("\nStep 6: Evaluation")

print("POS Tagging Completed Successfully.")

print("\nFinal Output")

for word, tag in pos_tags:
    print(f"{word:15} {tag}")
```

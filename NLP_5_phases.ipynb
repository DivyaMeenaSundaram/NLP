```python
# 5 Phases of NLP - Beginner Demo

sentence = "The cat chased the mouse."

print("Original Sentence:")
print(sentence)

# 1. Lexical & Morphological Analysis
words = sentence.lower().replace(".", "").split()
print("\n1. Lexical & Morphological Analysis")
print("Words:", words)

# 2. Syntactic Analysis
print("\n2. Syntactic Analysis")
print("Subject : cat")
print("Verb    : chased")
print("Object  : mouse")

# 3. Semantic Analysis
print("\n3. Semantic Analysis")
print("Meaning: A cat is chasing a mouse.")

# 4. Discourse Integration
previous = "The cat was hungry."
print("\n4. Discourse Integration")
print("Previous Sentence:", previous)
print("Current Sentence :", sentence)
print("Combined Meaning : The hungry cat chased the mouse.")

# 5. Pragmatic Analysis
print("\n5. Pragmatic Analysis")
print("Real-world Understanding:")
print("The cat chased the mouse because it wanted food.")
```

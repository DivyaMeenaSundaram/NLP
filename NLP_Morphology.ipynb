```python
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
def stem_words(words):
    ps = PorterStemmer()
    return [ps.stem(word) for word in words]
```

```python
def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ, "N": wordnet.NOUN, "V": wordnet.VERB, "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)
```

```python
def lemmatize_words(words):
    lemmatizer = WordNetLemmatizer()
    return [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in words]
```

```python
def preprocess_text(text):
    return word_tokenize(text.lower())
```

```python
text = "Dr. John Smith travelled from Chennai to Bengaluru to present his research paper. He met several professors after the presentation. They discussed future collaborations and planned to submit a joint research proposal."
words = preprocess_text(text)
stemmed_words = stem_words(words)
lemmatized_words = lemmatize_words(words)
print("Original Words:", words)
print("Stemmed Words:", stemmed_words)
print("Lemmatized Words:", lemmatized_words)
```
